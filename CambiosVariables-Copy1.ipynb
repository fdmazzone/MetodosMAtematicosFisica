<h2> Laplaciano en coordenadas polares </h2>

```python
x,y=symbols('x,y')
r=sqrt(x**2+y**2)
theta=atan(y/x)

 
```

```python
u=Function('u')(r,theta)
v=simplify(u.diff(x,2)+u.diff(y,2))
v
```

```python
r=symbols('r',positive=True)
theta=symbols('theta',real=True)
w=v.subs({sqrt(x**2+y**2):r,atan(y/x):theta})
w.simplify()
```

```python
w.subs({x:r*cos(theta),y:r*sin(theta)}).expand().simplify()
```

<h2>
    Encontrar el Laplaciano en coordenadas esféricas
</h2>


![](esfericas.png)

$$
x=r \sin \theta \cos \varphi \quad y=r \sin \theta \sin \varphi \quad z=r \cos \theta
$$

```python
var('x,y,z')
rho=sqrt(x**2+y**2)
r=sqrt(x**2+y**2+z**2)
theta=atan(rho/z)
phi=atan(y/x)
u=Function('u')(r,theta,phi)
v=u.diff(x,2)+u.diff(y,2)+u.diff(z,2)
v1=v.simplify().factor()
v1
```

```python
r=symbols('r',positive=True)
theta=symbols('theta',real=True)
phi=symbols('phi',real=True)
v2=v1.subs({x**2+y**2+z**2:r**2,atan(sqrt(x**2+y**2)/z):theta,atan(y/x):phi,sqrt(x**2+y**2):r*sin(theta)})
v2
```

```python
v3=v2.subs({x:r*sin(theta)*cos(phi),y:r*sin(theta)*sin(phi),z:r*cos(theta)})
v3.expand().simplify().expand().simplify().expand()
```

<h2> Cambio de variables analítico </h2>

**Teorema**
Supongamos $f(z)=u(x,y)+iv(x,y)$  analítica y $H(u,v)$ una función de clase $C^2$. Pongamos $h(x,y)=H(u(x,y),v(x,y))$ entonces
$$h_{xx}+h_{yy}=(H_{uu}+H_{vv})(v_x^2+v_y^2).$$
En paticular, si $f$ es mapeo conforme (es decir $f$ analítica y $f'(z)\neq 0$), entonces $H$ es armónica si y solo si $h$ es analítica.


```python
var('x,y')
u=Function('u')(x,y)
v=Function('v')(x,y)
H=Function('H')(u,v)
D2H=(H.diff(x,2)+H.diff(y,2)).simplify()
D2H.simplify()
```

```python
D2H.subs({u.diff(x):v.diff(y),u.diff(y):-v.diff(x),u.diff(x,2):-u.diff(y,2),v.diff(x,2):-v.diff(y,2) })
```
