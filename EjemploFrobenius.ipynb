<h3> Ejemplo 6.5.4 Aguirregabiria </h3>

$$(x+x^2)y''-xy'+y=0$$

```python
x=symbols('x',real=True)
a=symbols('a:10',real=True)
m,b=symbols('m,b',real=True)
y=sum(a[i]*x**(m+i) for i in range(10))
y
```

```python
eq=(x+x**2)*y.diff(x,2)-x*y.diff(x)+y
eq.expand().simplify().collect(x) #factor, 
```

```python
eq1=(eq/x**(m-1)).expand().collect(x)
eq1
```

```python
eq1.subs(x,0)
```

```python
solve(eq1.subs(x,0),m)
```

```python
eq2=eq1.subs(m,1)
eq2
```

```python
Ecuaciones=[eq2.coeff(x**i) for i in range(1,10)]
Ecuaciones
```

```python
a[1:]
```

```python
solve(Ecuaciones,a[1:])
```

Las ecuaciones son 
$$(m+n+1)a_n+a_{n+1}(m+n+1)^2=0$$

$$a_{n+1}=-\frac{a_n}{m+n+1}$$

Luego 
$$
a_{n}(m)=-\frac{a_{n-1}(m)}{m+n}=(-1)^{n} \prod_{k=1}^{n} \frac{1}{m+k}
$$
con $a_{0}(m)=1$. Una solución se obtiene tomando $m=0$ en la última expresión
$$
a_{n}(0)=\frac{(-1)^{n}}{n !}
$$
con lo que
$$
y_{1}=\sum_{n=0}^{\infty} \frac{(-1)^{n}}{n !} x^{n}=e^{-x}
$$
Para hallar la segunda solución mediante es conveniente utilizar la derivada logarítmica:
$$
\begin{aligned}
b_{n}=a_{n}^{\prime}(0) &=\left.a_{n}(0) \frac{d}{d m} \ln \left|a_{n}(m)\right|\right|_{m=0}=-\left.\frac{(-1)^{n}}{n !} \frac{d}{d m} \sum_{k=1}^{n} \ln (m+k)\right|_{m=0} \\
&=-\left.\frac{(-1)^{n}}{n !} \sum_{k=1}^{n} \frac{1}{m+k}\right|_{m=0}=-\frac{(-1)^{n}}{n !} H_{n}
\end{aligned}
$$
donde hemos utilizados los números armónicos $H_{n}=1+1 / 2+1 / 3+\cdots+1 / n$. La solución general de la ecuación será, por tanto,
$$
y=C_{1} e^{-x}+C_{2}\left[e^{-x} \ln x-\sum_{n=1}^{\infty} \frac{(-1)^{n}}{n !} \Omega_{n} x^{n}\right]
$$
