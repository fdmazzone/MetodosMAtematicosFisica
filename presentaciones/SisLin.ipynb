<h1>La matriz fundamental</h1>

 **Teorema: Existencia y unicidad para Sistemas Lineales** Sea $A:I \rightarrow \mathbb{R}^{n \times n}$, $b: I \rightarrow \mathbb{R}^{n}$ continuas. Entonces para cada $x_{0} \in \mathbb{R}^{n}$ existe una √∫nica solucion $x: I \rightarrow \mathbb{R}^{n}$ del PVI.
$$
\left\{
\begin{array}{ll}
 x^{\prime}(t)&=A(t) x(t)+b(t), \\
 x\left(t_{0}\right)&=x_{0}
\end{array}
\right.
$$

 
 
**Definicion: Matriz fundamental** Sean $e^1, \ldots e^{n}$ los vectores can√≥nicos de $\mathbb{R}^{n}$, $t_0\in I\subset\mathbb{R}$. Definimos $\Phi^{0}, \ldots, \Phi^{n}:I \rightarrow \mathbb{R}^{n}$ como
las soluciones de los PVI
$$
\left\{
\begin{split}
{x}'(t)&=A(t)x(t) \\
x(t_{0})&=e^{k}
\end{split}
\right.
$$
La **matriz fundamental** se define como
$$
G=\left[\Phi^{1}|\ldots| \Phi^{n}\right]
$$


**Observaci√≥n.** La matriz fundamental satisface
$$
\left\{
\begin{split}
{G}'(t)&=A(t) G(t) \\
G(t_0)&=I_{n\times n}
\end{split}\right.
$$


**Ejemplo: Osciladar arm√≥nico.**
$$
{x}''+\omega^{2} x=0.
$$
Si $x_{1}=x, x_{2}=\dot{x}$

\begin{equation}
\left\{
\begin{split}
\dot{x}_{1}&=x_{2},\\
\dot{x}_2 &=-\omega^{2} x_1\\
\end{split}
\right.
\end{equation}

Sabemos que

\begin{align*}
 x_1(t)&=x_1^0\cos \omega t+x_2^0\operatorname{sen}\omega t,\\
 x_2(t)&=- x_1^0\omega\operatorname{sen} \omega t+x_2^0\omega\cos\omega t,
 \end{align*}
 
Luego

$$
\left(\begin{array}{l}
x_{1} \\
x_{2}
\end{array}\right)=
\left(\begin{array}{ll}
1 & 0 \\
0 & \omega
\end{array}\right)
\left(\begin{array}{cc}
\cos \omega t & \operatorname{sen} \omega t \\
-\operatorname{sen} \omega t & \cos \omega t
\end{array}\right) \left(\begin{array}{l}
x_{1}^{0} \\
x_{2}^{0}
\end{array}\right)
=\left(\begin{array}{ll}
1 & 0 \\
0 & \omega
\end{array}\right)   R(-\omega t)\left(\begin{array}{l}
x_{1}^{0} \\

\end{array}\right)
$$

Para encontrar la matriz fundamental hay que resolver las condiciones iniciales $x(0)=e^{1}$  y $x(0)=e^{2}$. La condici√≥n $x(0)=e^1 \Rightarrow x_{1}^{0}=1$ y $x_{2}^{0}=0$ y la condici√≥n
$x(0)=e^{2} \Rightarrow x_1^0=0$ y $x_0^{2}=1 / \omega$. Luego las columnas de la MF son

$$\Phi_{1}=\left(\begin{array}{ll}1 & 0 \\ 0 & \omega\end{array}\right) R(-\omega t)\left(\begin{array}{l}1 \\ 0\end{array}\right),\quad
\Phi_{2}=\left(\begin{array}{cc}1 & 0 \\ 0 & \omega\end{array}\right) R(-\omega t)\left(\begin{array}{l}0 \\ 1 / \omega\end{array}\right)
$$

Entonces 

\begin{equation}
G=\left[\Phi_{1} \mid \Phi_{2}\right]=\left(\begin{array}{ll}1 & 0 \\ 0 & \omega\end{array}\right) R(-\omega t)\left(\begin{array}{ll}1 & 0 \\ 0 & 1/\omega\end{array}\right)=H R(-\omega t) H^{-1}
\end{equation}


**Teorema.** si $G(t)$ es una matriz fundamental  entonces $\Phi(t)=G(t)x_0$, con $x_{0} \in \mathbb{R}^{n}$, es una soluci√≥n general del problema homog√©neo. 
$$x'=A x.$$

**Teorema.** Supongamos $t_0\in I\subset\mathbb{R}$, $x_0\in\mathbb{R}^n$, $A: I \rightarrow \mathbb{R}^{n\times n}$ y $b: I \rightarrow \mathbb{R}^{n}$ y que $G(t)$ matriz fundamental de $A(t)$ ($G'=AG$, $G(t_0)=I_{n\times n}$), entonces

$$
    \varphi(t)=G(t)\left[x_{0}+\int_{t_{0}}^{t}(G(s))^{-1} b(s) d s\right]

\begin{equation}

\right.
\end{equation}

$$
\begin{aligned}

$$
\varphi(t) =G(t) x_{0}+G(t) \int_{t_{0}}^{t}(G(s))^{-1} b(s) d s \\

Notar que $\varphi_h$ resuelve el problema homog√©neo y $\varphi_p$ el no-homog√©neo.

<h1>Sistemas lineales homog√©neos con coeficientes constantes</h1>

Son sistenas de la forma
$$
x'=Ax, \quad A \in \mathbb{R}^{n \times n}\text{ constante}
$$

La forma de resolver es considerar la sustituci√≥n

$$
y=P^{-1} x
$$

donde $P$ es no singular.  Entonces

$$
y'=P^{-1} x'=P^{-1} A x=P^{-1}A Py
$$

Es una ecuaci√≥n de la forma 
$$y'=By,\quad \text{con } B \sim A.$$

¬øQue sistemas sabr√≠amos resolver? Si $B$ fuese diagonal 

$$B=\operatorname{diag} \left\{ \lambda_{1},\ldots \lambda_{n}\right\}$$

Entonces quedar√≠a un sistema **desacomplado** que podr√≠amos resolver

$$
\begin{split}
y'_{1}&=\lambda_{1} y_{1} \\
&\quad \vdots\\
y'_{n}&=\lambda_{n} y_{n} \\
\end{split}
$$

Las soluciones son:
$$y_{j}(t)=A_{j} e^{\lambda_{j} t},\quad A_j\in\mathbb{R}$$
 
La matriz fundamental es
\begin{equation} 
G(t)=\begin{pmatrix}
e^{\lambda_{1} t} & \\
& \ddots \\
& e^{\lambda_{n} t}\\
\end{pmatrix}
\end{equation}

Pero desafortunadamente no toda matriz es diogonalizable.

<h2>Exponencial de matrices</h2>


La soluci√≥n de 

$$
\left\{
\begin{split}
    x(t)&=\lambda x(t)  \\
    x(0)&=x_0 \\
\end{split}
\right.
$$
con $\lambda,x_0 \in \mathbb{R}$, $x:\mathbb{R}\to\mathbb{R}$, es $x(t)=e^{\lambda t} a$. Razonando por analogia, la de

$$
 \begin{split}
  x'&=Ax\\
  x(0)&=x_0\in\mathbb{R}^n\\
 \end{split}
$$

deber√≠a ser 
$$x(t)=e^{A t}x_0$$

Para ello deber√≠amos darle sentido a la exponencial de una matriz. 

Recordemos que sobre el espacio de las matrices cuadradas se dice que una norma se llama \emph{norma matricial} si satisface
$$   
  \|AB\|\leq \|A\|\|B\|.
$$
Hay muchas normas matriciales, todas equivalentes. Por ejemplo
$$
 \|A\|=\sup\{\|Ax\| \mid  x\in\mathbb{R}^n, \|x\|=1\}.
$$

Estas normas dotan a $\mathbb{R}^{n\times n}$ de la noci√≥n de convergencia. Diremos que las matrices $A_n$ **convergen** a $A$ si
$$\|A_n-A\|\to 0.$$



**Definicion (Convergencia de series de matrices)** Sean $A_j\in\mathbb{R}^{n\times n}$, $j=0,1,\ldots$ una sucesi√≥n de matrices. Diremos que la serie 
$$
 \sum_{j=0}^{\infty} A_j
$$
converge si la sucesi√≥n de sumas parciales
$$
 S_m=\sum_{j=0}^{m} A_j
$$
 converge en la norma matricial. 
 
 
**Lema** El espacio de las matrices con la norma matricial satisface el criterio de Cauchy, Si $\{A_n\}$ es una sucesi√≥n tal que $\|A_n-A_m\|\to 0$ entonces existe una matriz $A$ tal que $A_n\to A$. 


**Teorema (Exponencial de matrices)** Sea $A\in\mathbb{R}^{n\times n}$. Entonces  la serie matricial
$$
\sum_{n=0}^{\infty} \frac{A^n}{n!}=I+A+\frac{A^{2}}{2 !}+\cdots
$$
converge. La matriz resultante se denomina exponencial de $A$ y se denota $e^A$.


**Demostraci√≥n.** Veamos que la sucesi√≥n de sumas parcales es de Cauchy. Sean $M>N$. 
$$
\left\|S_{M}-S_{N}\right\|=\left\|\sum_{j=N+1}^{M} \frac{A^{j}}{j!}\right\| \leqslant \sum_{j=N+1}^{M} \frac{\| A \|^j}{j !}
$$
 y la √∫ltima suma puede hacerse tan  chica como queramos pues la serie
$$
  e^{\|A\|}=\sum_{n=0}^{\infty}\frac{\|A\|^n}{n!}
$$
es convergente $\square$



**Ejemplo (Matriz diagonal)**  Si
 $$
 A=\begin{pmatrix}
\lambda_{1}&0 & \cdots \\
\vdots & \ddots& \vdots\\
0 &\cdots & \lambda_{n}\\
\end{pmatrix}
$$

Entonces 

 $$
 A^j=\begin{pmatrix}
\lambda^j_{1}&0 & \cdots \\
\vdots & \ddots& \vdots\\
0 &\cdots & \lambda^j_{n}\\
\end{pmatrix}
$$
Luego si $t\in\mathbb{R}$

 $$
 e^{tA}=\sum_{j=0}^{\infty}\frac{t^j}{j!}\begin{pmatrix}
\lambda^j_{1}&0 & \cdots \\
\vdots & \ddots& \vdots\\
0 &\cdots & \lambda^j_{n}\\
\end{pmatrix}=
\begin{pmatrix}
\sum_{j=0}^{\infty}\frac{t^j}{j!}\lambda^j_{1}&0 & \cdots \\
\vdots & \ddots& \vdots\\
0 &\cdots & \sum_{j=0}^{\infty}\frac{t^j}{j!}\lambda^j_{n}\\
\end{pmatrix}
=
\begin{pmatrix}
e^{\lambda_{1}t}&0 & \cdots \\
\vdots & \ddots& \vdots\\
0 &\cdots & e^{\lambda_n t}\\
\end{pmatrix}
$$
Que resulta ser la matriz fundamental del sistema $x'=Ax$.

3. Si $A$  conmuta con $B$ $(A B=B A)$ entonces

**Demostraci√≥n** 

Demostremos 2. Observar que
$$
 \left(P^{-1}AP\right)^n=P^{-1}APP^{-1}AP\cdots P^{-1}AP=P^{-1}A^nP
$$
Luego
$$e^{P^{-1} A P}= \sum_{n=0}^{\infty}\frac{ \left(P^{-1} A P\right)^n}{n!}
= \sum_{n=0}^{\infty}\frac{P^{-1} A^n P}{n!} 
=P^{-1} \left(\sum_{n=0}^{\infty}\frac{ A^n}{n!}  \right)P=P^{-1}e^AP.
$$


Demostremos 3. Usando la formula del binomio de Newton 
$$(A+B)^{n}=\sum_{j=0}^{n} \binom{n}{j} A^{j} B^{n-j}$$
Luego
$$
e^{A+B}=\sum_{n=0}^{\infty}\frac{1}{n!} \sum_{j=0}^{n}\binom{n}{j}
A^{j} B^{n-j}
$$
Por otro lado
$$
\begin{aligned}
e^{A}  e^{B}&=\left(I+A+\frac{A^{2}}{2 !}+\frac{A^{3}}{3 !}+\cdots\right)\left(I+B+\frac{B^{2}}{2}+\frac{B^{3}}{3}+\cdots\right) \\
&=I+A+B+\frac{A^{2}}{2}+A B+\frac{B^{2}}{2}+\frac{A^{3}}{3 !}+\frac{A^{2}}{2 !} B+A \frac{B^{2}}{2 !}+\frac{B^{3}}{3}+\cdots\\
&=\sum_{n=0}^{\infty} \frac{1}{n!}\sum_{j=0}^{n}\binom{n}{j} A^j B^{n-j}=e^{A+B}
\end{aligned}
$$
 $\square$

 $$

<h2>Exponenciales y formas de Jordan</h2>

Podemos usar las formas de Jordan Para calcular de manera expl√≠cita la exponencial de una matriz. 

Un bloque elemental  de Jordan asociado a un autovalor real $\lambda\in\mathbb{R}$ es una matriz  de la siguiente forma
$$
\begin{split}
 J&=
   \begin{pmatrix}
    \lambda & 1        & \cdots& \cdots  &0\\
    0         &\lambda  &  1      &\cdots &0 \\
    \vdots    & \vdots   &\ddots  &\ddots&\vdots \\
     \vdots    & \vdots  & \vdots   &\lambda  &1\\
      0    & 0   & \cdots& \cdots &\lambda\\   
   \end{pmatrix}\\
   &=
      \begin{pmatrix}
    \lambda & 0        & \cdots& \cdots  &0\\
    0         &\lambda  &  0      &\cdots &0 \\
    \vdots    & \vdots   &\ddots  &\ddots&\vdots \\
     \vdots    & \vdots  & \vdots   &\lambda  &0\\
      0    & 0   & \cdots& \cdots &\lambda\\   
   \end{pmatrix}
  +
      \begin{pmatrix}
    0 & 1        & \cdots& \cdots  &0\\
    0         &0  &  1      &\cdots &0 \\
    \vdots    & \vdots   &\ddots  &\ddots&\vdots \\
     \vdots    & \vdots  & \vdots   &0  &1\\
      0    & 0   & \cdots& \cdots &0\\   
   \end{pmatrix}\\
   &=\lambda I+N.
\end{split}
$$
Como $t\lambda I$ y $tN$ conmutan
$$e^{tJ}=e^{t(\lambda I+N)}=e^{t\lambda I}e^{tN}.$$
Cada uno de los factores $e^{t\lambda I}$ y $e^{tN}$ los hemos calculado. 


Si $J$ esta compuesta de varios bloques elementales de Jordan

$$
 J=
   \begin{pmatrix}
    J_1 & 0        & \cdots& \cdots  &0\\
    0         &J_2  &  0      &\cdots &0 \\
    \vdots    & \vdots   &\ddots  &\ddots&\vdots \\
     \vdots    & \vdots  & \vdots   &J_{k-1}  &0\\
      0    & 0   & \cdots& \cdots &J_k\\   
   \end{pmatrix}
$$

Entonces 
$$
 e^{tJ}=
   \begin{pmatrix}
    e^{tJ_1} & 0        & \cdots& \cdots  &0\\
    0         &e^{tJ_2}  &  0      &\cdots &0 \\
    \vdots    & \vdots   &\ddots  &\ddots&\vdots \\
     \vdots    & \vdots  & \vdots   &e^{tJ_{k-1}}  &0\\
      0    & 0   & \cdots& \cdots &e^{tJ_k}\\   
   \end{pmatrix}
$$

Utilizando estos recursos podemos calcular la exponencial de cualquier matriz de la que conozcamos su forma de Jordan. 

<h2> Caso autovalores complejos </h2>
Cuando hay autovalores complejos los resultados de los calculos anteriores nos llevan a soluciones  que toman valores complejos. Sin embargo, suele ser el caso que busquemos soluciones que tomen valores reales.

Es relativamente sencillo obtener soluciones reales a partir de soluciones complejas, veamos un ejemplo particular que va a ilustrar como proceder en general.

Supongamos $\lambda\in\mathbb{C}\notin\mathbb{R}$ un autovalor complejo y no real y $P\in\mathbb{C}^n$ su correspondiente autovector. Por un ejercicio de m√°s arriba sabemos que $\varphi(t)=e^{\lambda t}P$ es soluci√≥n con valores en $\mathbb{C}$.  Escribamos
$$
\lambda=\mu+i\nu,\quad P=Q+iR, \quad \mu,\nu\in\mathbb{R}\hbox{ y } Q,R\in\mathbb{R}^n.
$$

Como la matriz $A$ es real $\bar{\lambda}=\mu-i\nu$ es autovector y $\bar{P}=Q-iR$ es un autovector correspondiente y tambien se tiene que $\bar{\varphi}e^{\bar{\lambda}t}\bar{P}$ es soluci√≥n. Luego  las siguientes  combinaciones lineales son soluci√≥n

$$
\psi_1(t)=\frac12\varphi(t)+ \frac12\bar{\varphi}(t)=\boxed{e^{\mu t}\left(\cos\nu t Q-\operatorname{sen}\nu t R   \right)}\\
\psi_2(t)=-\frac{i}{2}\varphi(t)+ \frac{i}{2}\bar{\varphi}(t)=\boxed{e^{\mu t}\left(\cos\nu t R+\operatorname{sen}\nu t P   \right)}
$$
es soluci√≥n. Se puede demostrar que son linealmente independientes en $\mathbb{R}$.  

**Conclusi√≥n** Cuando tenemos un autovalor complejo, lo juntamos con su conjugado y formamos el doble de soluciones. Del par de autovalores conjugados solo consideramos uno. 

Cuando hay autovalores deficientes se procede de manera similar.

<h2> El problema no-homog√©neo para sistemas con coeficientes constantes</h2>

**Corolario.** Sea $A\in \mathbb{R}^{n\times n}$ constante, $t_o\in\mathbb{R}$ y $x_0\in\mathbb{R}^n$. La soluci√≥n del PVI
$$
 \left\{
    \begin{array}{ll}
        x^{\prime}&=A x+b(t)\\
        x(t_0)&=x_0\\
    \end{array}
  \right.
$$
viene dada por la expresi√≥n

$$x(t)=e^{A(t-t_0)} x_0+\int_{t_0}^{t} e^{A(t-s)} b(s) d s.$$


**Demostraci√≥n**  Se verifica que la matriz  $G(t)=e^{A(t-t_0)}$ resuelve $G'(t)=AG(t)$ y $G(t_0)=I$. Luego es una matriz fundamental del sistema $x'=Ax$ en $t_0$. Usando la f√≥rmula para resolver sistemas no homog√©neos deuducimos
$$
 x(t)=e^{A(t-t_0)}\left[x_0 +\int_{t_0}^t e^{A(t_0-s)}b(s)ds\right]=e^{A(t-t_0)}x_0 +\int_{t_0}^t e^{A(t-s)}b(s)ds
$$

$\square$
