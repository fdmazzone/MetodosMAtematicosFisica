<h1> Combinando Cosas </h1>

```python
from scipy.integrate import odeint
```

```python
def f(z,t,omega):
    x,v=z
    return [v,-omega**2*x]
    
```

```python
z0=[1.0,0.0]
t=np.linspace(0,10, 1000)
sol=odeint(f,z0,t,args=(1.,))
```

```python
np.shape(sol)
```

```python
sol[:3,:]
```

```python
plt.plot(t,sol[:,0])
```

```python
plt.plot(t,sol)
```

```python
x=sol[:,0]
v=sol[:,1]
plt.plot(x,v)
```

```python
fig,ax=plt.subplots()
t=np.linspace(0,10, 1000)
for z0 in  np.arange(0,5,.1):
    z=[z0,0.0]
    sol=odeint(f,z,t,args=(1.,))
    x=sol[:,0]
    v=sol[:,1]
    ax.plot(x,v)
```
